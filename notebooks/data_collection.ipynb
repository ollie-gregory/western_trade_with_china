## Global Emissions Data

This data is downloaded from the European Commission website, available [here](https://edgar.jrc.ec.europa.eu/report_2024).

Crippa, M., Guizzardi, D., Pagani, F., Banja, M., Muntean, M., Schaaf, E., Monforti-Ferrario, F., Becker, W.E., Quadrelli, R., Risquez Martin, A., Taghavi-Moharamli, P., Köykkä, J., Grassi, G., Rossi, S., Melo, J., Oom, D., Branco, A., San-Miguel, J., Manca, G., Pisoni, E., Vignati, E. and Pekar, F., GHG emissions of all world countries, Publications Office of the European Union, Luxembourg, 2024, doi:10.2760/4002897, JRC138862

```python
emissions_data = pd.read_excel('../data/unprocessed/global_emissions.xlsx', sheet_name='fossil_CO2_by_sector_country_su')
emissions_data = emissions_data.drop(columns=['Substance', 'EDGAR Country Code'])
emissions_data.head()
```

```python
# Melt the data so that each row is a unique year, country, sector combination
emissions_data_melted = emissions_data.melt(id_vars=['Sector', 'Country'], var_name='Year', value_name='Emissions')

# Reorder the columns to be year, country, sector, emissions
emissions_data_melted = emissions_data_melted[['Year', 'Country', 'Sector', 'Emissions']]
emissions_data_melted.columns = ['year', 'country', 'sector', 'emissions']

emissions_data_melted.head()
```

```python
# Export the data to a csv file
emissions_data_melted.to_csv('../data/processed/global_emissions.csv', index=False)
```
